# GREAT

Analyze regions on the web tool, use the whole genome option as the background for all conditions analyzed. Also use the Basal plus extension method.  

Took top 5 hits by hyper P and put into excel spreadsheet and prepped for plotting, then loaded back into ACCRE. 

```r
suppressPackageStartupMessages(library(tidyverse))

res <- read_tsv('/data/hodges_lab/ATAC-STARR_B-cells/results/results_human-evolution/GREAT/GO-BP_processed_results_plot-ready.txt', 
                 col_names = c("regionset", "term", "binom_q")) %>% mutate(logp = -log10(binom_q))
```

```r
options(repr.plot.width = 8, repr.plot.height = 2.5)

res_p <- filter(res, regionset == "conserved active") %>% mutate(term = fct_reorder(term, logp))

p <- ggplot(res_p, aes(y =logp, x = term)) +
    geom_bar(stat = "identity", color = "black", fill = "white", lwd = 0.2, width = 0.8) +
    coord_flip() +
    facet_wrap(~regionset, scales = "free", ncol = 1) +
    theme_minimal() +
    theme(legend.position = "none", axis.title.y = element_blank())

p

ggsave(p, filename = "/data/hodges_lab/ATAC-STARR_B-cells/results/results_human-evolution/GREAT/cons_act_GO-bp.pdf", 
    units = c("in"), height = 2.5, width = 8, device = cairo_pdf)
```
