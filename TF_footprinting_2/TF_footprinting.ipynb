# Transcription Factor Footprinting

## GM12878 TF footprinting

```bash
# bash kernel 

# assign variables
hg38='/data/hodges_lab/hg38_genome/hg38.fa'
PREFIX='GM12878_1.8-filter'
MOTIF='/home/hansetj1/JASPAR2020_CORE_vertebrates_non-redundant_pfms_jaspar.txt'

#from ini
BAM="$(read-ini ../fantom/config.ini BAM merged_path)/GM12878inGM12878_DNA_merged-3reps.unique.pos-sorted.bam"
PEAKS=$(read-ini ../fantom/config.ini CHRACC gg_peaks)
OUT=$(read-ini ../fantom/config.ini TF_FOOTPRINTING path)
```

### ATACorrect

```bash
TOBIAS ATACorrect --bam $BAM \
    --genome $hg38 \
    --peaks $PEAKS \
    --outdir ${OUT}/ATACorrect \
    --cores 8 --prefix $PREFIX
```

### ScoreBigwig

```bash
TOBIAS ScoreBigwig --signal ${OUT}/ATACorrect/${PREFIX}_corrected.bw \
    --regions $PEAKS \
    --output ${OUT}/ScoreBigwig/${PREFIX}_footprints.bw --cores 8
```

### BINDetect

```bash
TOBIAS BINDetect --bound-pvalue 0.05 --motifs $MOTIF \
    --signals ${OUT}/ScoreBigwig/${PREFIX}_footprints.bw \
    --genome $hg38 --peaks $PEAKS \
    --outdir ${OUT}/BINDetect/${PREFIX}_0.05-pval --cond_names $PREFIX --cores 8
```

```bash
#Archetypes

MOTIF='/home/hansetj1/consensus_pwms.meme'

TOBIAS BINDetect --bound-pvalue 0.05 --motifs $MOTIF \
    --signals ${OUT}/ScoreBigwig/${PREFIX}_footprints.bw \
    --genome $hg38 --peaks $PEAKS \
    --outdir ${OUT}/BINDetect/${PREFIX}_archetypes_0.05-pval --cond_names $PREFIX --cores 4
```

## Compare bound and unbound sites in profiles

For a supplementary figure, we want to indicate that we can do TF footprinting in both genomes. The best way I can think of showing this is profile plots for unbound and bound TFs. 

### Human

```bash
#BedFiles
BED_DIR='/data/hodges_lab/ATAC-STARR_B-cells/results/ATAC-STARR_TF-footprinting/BINDetect/GM12878_1.8-filter_0.05-pval'

#BigWigs
CC_SIGNAL='/data/hodges_lab/ATAC-STARR_B-cells/results/ATAC-STARR_TF-footprinting/ATACorrect/GM12878_1.8-filter_corrected.bw'

#Output
OUTPUT_DIR='/data/hodges_lab/ATAC-STARR_B-cells/results/results_human-evolution/TF_footprinting/plots'

for i in ETS1_MA0098.3 ELK3_MA0759.1 PRDM1_MA0508.3 #IRF4_MA1419.1 SPI1_MA0080.5 NFKB1_MA0105.4 
do
    computeMatrix reference-point -S $CC_SIGNAL \
        -R ${BED_DIR}/${i}/beds/${i}_GM12878_1.8-filter_bound.bed \
        ${BED_DIR}/${i}/beds/${i}_GM12878_1.8-filter_unbound.bed \
        -a 75 -b 75 --referencePoint center --missingDataAsZero --binSize 1 -p 4 \
        -o ${OUTPUT_DIR}/matrix_footprinting_${i}.gz

    plotProfile -m ${OUTPUT_DIR}/matrix_footprinting_${i}.gz -o ${OUTPUT_DIR}/aggregate_footprinting_${i}.pdf \
        --dpi 300 --plotFileFormat pdf --colors black grey \
        --refPointLabel center --yAxisLabel "Accessible TF Motifs" --regionsLabel "bound motif" "unbound motif"\
        --samplesLabel "Human Cut-Counts" --plotWidth 10 --plotHeight 8 #in cm
done
```
